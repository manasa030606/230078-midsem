# Task 1.2 Key Assumptions

## Assumption 1
**Assumption:** The classifier weights $w$ follow a multivariate Gaussian distribution $w \sim \mathcal{N}(\mu, \Sigma)$.

**Why the method needs it:** The proposed contribution tracks the uncertainty/confidence of each feature's weight. The Gaussian assumption allows the algorithm to express the constraint on the probability of correct classification (i.e., $\text{Pr}[y_t(w \cdot x_t) \ge 0] \ge \eta$) compactly in a closed-form deterministic margin formulation involving the mean $\mu$ and variance terms from $\Sigma$.

**Violation scenario:** In a scenario where the true optimal weights lie in a multi-modal distribution or are strictly uniform within certain bounds (e.g., highly disjoint sub-populations with completely different decision boundaries that cannot be captured by a single Gaussian mean), representing weights with a single Gaussian would fail to capture the true underlying structure.

**Paper reference:** Section 2.3 (Background on CW Learning), Section 3 ("we assume the weight vector w follows the Gaussian distribution with the mean vector $\mu$ and the covariance matrix $\Sigma$").

---

## Assumption 2
**Assumption:** The cost of misclassification can be traded off linearly (or squared) against the change in the weight distribution using a single fixed hyperparameter $C$.

**Why the method needs it:** To achieve the "soft" margin property, the algorithm formulates the optimization (Eq. 4 and Eq. 5) by adding the confidence loss scaled by $C$ directly to the KL divergence penalty. This enables the method to limit the aggressiveness of the update when faced with noisy or mislabeled instances that would otherwise cause huge shifts in the parameters.

**Violation scenario:** In a highly non-stationary environment (concept drift) where the level of noise or the penalty for misclassifying changes dynamically over time (e.g., streaming financial data during a market crash versus normal trading days), using a fixed scalar $C$ to trade-off passiveness over the entire sequence would be suboptimal.

**Paper reference:** Section 3, Equations 4 and 5 (optimization formulations showing the $+ C\ell_\phi$ and $+ C\ell_\phi^2$ terms).

## Assumption 3
**Assumption:** A linear combination of the input features $x$ is sufficient to distinguish the classes (i.e., the decision boundary is inherently linear in the feature space, or the kernelized feature space).

**Why the method needs it:** The prediction rule $\hat{y}_t = \text{sgn}(\mu_{t-1} \cdot x_t)$ and the loss function formulate the margin $y_t(\mu \cdot x_t)$ as a linear dot product. The confidence-weighted updates strictly operate on the assumption that we are adjusting a hyperplane separated by a linear margin.

**Violation scenario:** A dataset with a pure XOR structure or concentric circles where the classes are inextricably intertwined non-linearly, and no kernelization is applied prior to feeding the features into the standard linear SCW algorithm.

**Paper reference:** Section 2.1 (Overview of Online Learning defining model prediction) and Section 3 (defining the loss function $\ell_\phi(\mathcal{N}(\mu_{t-1}, \Sigma_{t-1}); (x_t, y_t)) = \max(0, \phi \sqrt{x_t^\top \Sigma_{t-1} x_t} - y_t \mu_{t-1} \cdot x_t)$, where the margin is purely linear).
